```python
# import libraries
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in data to dataframe, remove MYOPIC column from the dataset
df = pd.read_csv("myopia.csv")
target = df.MYOPIC
df.drop(columns=["MYOPIC"], inplace=True)
df.head()
```

```python
df.info()
```

```python
# Code structure for t-SNE plotting borrowed from in-class lecture on this topic.

X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1])
plt.title("t-SNE Scatter Plot")
plt.show()
```

```python
# Code structure for PCA model borrowed from in-class lecture on this topic.

pca = PCA()
pca.fit(df)

plt.plot(range(len(df.columns)), pca.explained_variance_ratio_)
plt.xlabel("# of Principal Components")
plt.ylabel("Explained Variance")
plt.title("PCA Elbow")
plt.show()
```

```python
# Instructor's idea to list out components of predictiveness of columns in dataset to understand further which three components
# control about 90% of the model.

list(zip(range(len(df.columns)), pca.explained_variance_ratio_))
```

```python
# Instructor's idea to examine correlations of components within dataset.

df.corr()
```

```python
# Instructors idea to separate the top three PCA components based on the PCA Model.

pca = PCA(n_components=3)
pca.fit(df)

components = pd.DataFrame(pca.transform(df), columns=["PCA1", "PCA2", "PCA3"])

components.head()
```

```python
# Code structure for K-means cluster analysis borrowed from in-class lecture on this topic.

inertia = []
k_range = list(range(2, 10))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    
    inertia.append(kmeans.inertia_)
    
# make plot
plt.figure(figsize=(10,6))

plt.plot(k_range, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")

plt.show()
```

Elbow for the plot above appears to be at a k value of 4 (middle of curve).
